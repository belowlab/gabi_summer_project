```python
# -------- Helper function --------
def median(vals):
    '''
    Parameter: vals is an array of values. Strings and missing values are allowed
    Return: median of vals
    '''
    # Convert values to numbers, handle missing values
    cleaned_vals = []
    for val in vals:
        try: cleaned_vals.append(float(val))
        except: pass
    
    # Calcualte median
    cleaned_vals.sort()
    med = (cleaned_vals[len(cleaned_vals)//2] + cleaned_vals[(len(cleaned_vals)-1)//2])/2
    return med
# -------- End --------

lst_amygdala, lst_hippocampusm, lst_nucAccumbens, lst_peduncle, lst_fornixCres = [], [], [], [], []
with open('PythonClass_Final_Dataset.csv') as fh:
    line = fh.readline() # Skip header
    line = fh.readline().strip()
    while line != '':
        tmp = line.split(',')
        lst_amygdala.append(tmp[1])
        lst_hippocampusm.append(tmp[2])
        lst_nucAccumbens.append(tmp[3])
        lst_peduncle.append(tmp[4])
        lst_fornixCres.append(tmp[5])
        line = fh.readline().strip()

print('Amygdala:', median(lst_amygdala))
print('Hippocampusm:', median(lst_hippocampusm))
print('NucAccumbens:', median(lst_nucAccumbens))
print('Peduncle:', median(lst_peduncle))
print('FornixCres:', median(lst_fornixCres))
```

### 2. Pandas

```python
import pandas as pd
df = pd.read_csv('PythonClass_Final_Dataset.csv')
df.iloc[:, 1:].apply(func=median) # Skip the first column

# Or use aggragate()
# df.iloc[:, 1:].aggregate(func=median)
```

### 3. Plot

```python
df.hist(rwidth=0.9, )
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(12, 6), dpi=100)
fig.delaxes(ax[1,2])
tissues = ['Tissue_Amygdala', 'Tissue_Hippocampus', 'Tissue_NucAccumbens', 'Tissue_Peduncle', 'Tissue_FornixCres']
row, col = 0, 0
for tissue in tissues:
    ax[row, col].hist(df[tissue], rwidth=0.9)
    ax[row, col].set_xlabel('CpG Site values')
    ax[row, col].set_ylabel('Counts')
    ax[row, col].set_title(f'{tissue}')
    if col == 2:
        row += 1
        col = 0
    else:
        col += 1

fig.tight_layout()
```
